```python
faq = {
    "What is AI?": "AI stands for Artificial Intelligence.",
    "Who created Python?": "Python was created by Guido van Rossum.",
    "What is machine learning?": "Machine learning is a branch of AI that enables systems to learn from data.",
    "What is CodeAlpha?": "CodeAlpha is a software development company offering internships."
}

questions = list(faq.keys())
answers = list(faq.values())

print("FAQ Data Loaded Successfully")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def get_response(user_question):
    all_questions = questions + [user_question]

    vectorizer = CountVectorizer()
    vectors = vectorizer.fit_transform(all_questions)

    similarity = cosine_similarity(vectors[-1], vectors[:-1])
    best_match = similarity.argmax()

    return answers[best_match]
```

```python
user_question = input("Ask your question: ")

response = get_response(user_question)

print("Bot:", response)
```

```python
def get_response(user_question):
    all_questions = questions + [user_question]

    vectorizer = CountVectorizer()
    vectors = vectorizer.fit_transform(all_questions)

    similarity = cosine_similarity(vectors[-1], vectors[:-1])

    best_match = similarity.argmax()
    score = similarity[0][best_match]

    if score < 0.3:
        return "Sorry, I don't understand that question."

    return answers[best_match]
```

```python
import ipywidgets as widgets
from IPython.display import display

question_box = widgets.Text(
    placeholder='Ask something...',
    description='You:'
)

button = widgets.Button(description="Ask")
output = widgets.Output()

def chatbot(b):
    with output:
        output.clear_output()
        question = question_box.value
        response = get_response(question)
        print("Bot:", response)

button.on_click(chatbot)

display(question_box, button, output)
```
